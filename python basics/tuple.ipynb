```python
## nested tuples
lists = []
for i in range(1, 11):
    lists.append((i, i*5))
nested_tuple = tuple(lists)
nested_tuple
```

```python
for items in nested_tuple:
    for item in items:
        print(item, end=" ")
    print()
```
