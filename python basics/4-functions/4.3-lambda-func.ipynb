```python
#syntax

lambda arguments: expression
```

```python
def addition(a, b):
    return a + b

print(addition(5, 10))
```

```python
add = lambda a,b : a+b
print(add(5, 10))
type(add)
```

```python
def even_odd(num):
    if num % 2 == 0:
        return True

print(even_odd(4))
```

```python
even = lambda num: num%2 == 0
print(even(4))
print(even(5))
```
