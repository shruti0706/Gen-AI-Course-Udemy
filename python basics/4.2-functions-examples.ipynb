### 1. temperature converion

```python
#temp conevertor

def convert_temp(temp, unit):
    """this function converts the temperature from celsius to fahrenheit and vice versa"""
    if unit == "C":
        return (temp - 32) * 5/9
    elif unit == "F":
        return (temp * 9/5) + 32
    else:
        return "Invalid unit"

print(f"{convert_temp(100, "C") : .2f}")
print(f"{convert_temp(212, "F") : .2f}")
```

### 2. password strength checker

```python
def check_strong_pswd(pswd):
    """this function checks if the password is strong or not"""
    if len(pswd) < 8:
        return "Weak password"
    elif not any(char.isdigit() for char in pswd):
        return "Weak password"
    elif not any(char.isupper() for char in pswd):
        return "Weak password"
    elif not any(char.islower() for char in pswd):
        return "Weak password"
    elif not any(char in "!@#$%^&*()-_=+[]{}|;:'\",.<>?/`~" for char in pswd):
        return "Weak password"
    else:
        return "Strong password"
    
print(check_strong_pswd("Password123!"))
print(check_strong_pswd("password"))
# Weak password (no uppercase, digit, special character)

print(check_strong_pswd("Password"))
# Weak password (no digit, special character)

print(check_strong_pswd("Password1"))
# Weak password (no special character)

print(check_strong_pswd("Pass1!"))
# Weak password (length < 8)

print(check_strong_pswd("Password123!"))
# Strong password
```

### 3. calculate total cost of items in a shopping cart

```python
# 3. calculate total cost of items in a shopping cart
cart = [
    {"name ": "apple", "quantity" : 4, "price" : 5 },
    {"name ": "banana", "quantity" : 2, "price" : 4 },
    {"name ": "oranges", "quantity" : 2, "price" : 5 }
]

def calc_total(cart):
    total = 0
    for item in cart:
        total += item["quantity"] * item["price"]
    return total

print(calc_total(cart))
```

### 4.check if string is palindrome or not

```python
# str = "hello world"
str = "edryrde"
length = len(str)
mid = int(length/2)


def check_palindrome(str):  #this only words for a single word
    for i in range(0, mid):
        if str[i] != str[length - i- 1]:
            return "no"
    return "yes"

sentence = "A man a plan a canal Panama"

def is_palindrome(str): # this works for a sentence as well
    str = str.lower().replace(" ","")
    return str == str[::-1]


print("is it a palindrome ? --", is_palindrome(sentence))
```

### 5. Calculate factorials of a number using recursion

```python
def fact(num):
    if num == 0 or num == 1:
        return 1
    return num * fact(num-1)

print(fact(5))
```
