```python
# read whole file

with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip()) #strip removes the extra new line character at the end of each line.
```

```python
# write a file(overwriting)

with open('example.txt', 'w') as file:
    file.write('heeyyyyyy!!\n')
    file.write('this is new line.')
```

```python
# write a file( without overwriting )

with open('example.txt', 'a') as file:
    file.write('adding a new append text.\n')
```

```python
# writing lines in a files
lines = ['first line \n', 'second line \n', 'third line\n']

with open('example.txt', 'a') as file:
    file.writelines(lines)
```

```python
# write in binary file

data = b'\x00\x01\x02\x03\x04\x05'

with open('example.bin', 'wb') as file:
    file.write(data)    
```

```python
# read from binary file

with open('example.bin', 'rb') as file:
    binary_data = file.read()
    print(binary_data)
```

```python
# read the content from one file and write it to another file

with open('example.txt', 'r') as source_file:
    content = source_file.read()

with open('copy_example.txt', 'w') as destination_file:
    destination_file.write(content)
```

```python
# read a text file and count number of lines, words and characters

with open('example.txt', 'r') as file:
    content = file.read()
    lines = content.splitlines()
    words = content.split()
    characters = len(content)

    print(f'Number of lines: {len(lines)}')
    print(f'Number of words: {len(words)}')
    print(f'Number of characters: {characters}')
    print(f'Number of characters (excluding spaces): {characters - content.count(" ")}')
```

```python
# writing and then reading a file

with open('example.txt', 'w+') as file:
    file.write('This is a test file.\n')
    file.write('It contains some text.\n')
    file.write('We will read it back.\n')

    # Move the cursor to the beginning of the file
    file.seek(0)

    # Read the content of the file
    content = file.read()
    print(content)
```
