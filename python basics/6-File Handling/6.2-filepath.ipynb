```python
import os
cwd = os.getcwd()
print("Current working directory:", cwd)
```

```python
#create a new directory
new_directory = "package"
os.mkdir(new_directory)
print(f" new directory '{new_directory}' created.")
```

```python
# listing files and directories

items = os.listdir('.') 
print(items)
print(sorted(items))  ## to get items in sorted order 
```

```python
# joining paths

dir_name = 'folder'
file_name = 'file.txt'
full_path = os.path.join(dir_name, file_name)
print(full_path)
```

```python
 # joining paths

dir_name = 'folder'
file_name = 'file.txt'
full_path = os.path.join(os.getcwd(), dir_name, file_name)
print(full_path)
```

```python
# checking if file path exists or not
path1 = 'example.txt'

if os.path.exists(path1):
    print(f" {path1} exists.")
else:
    print(f" {path1} does not exists.")
```

```python
# checking if a path is a file or directory
path2 = 'example.txt'
if os.path.isfile(path2):
    print(f" {path2} is a file.")
else:
    print(f" {path2} is not a file.")
```

```python
## getting the absolute path of a file or directory
path3 = 'example.txt'
absolute_path = os.path.abspath(path3)
print(f"Absolute path of {path3} is: {absolute_path}")

relative_path = os.path.relpath(absolute_path)
print(f"Relative path of {path3} is: {relative_path}")
```
