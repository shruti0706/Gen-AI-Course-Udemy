```python
num = 24
def even_odd(num):
    """this function checks if a number is even or odd"""
    if num % 2 == 0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")


even_odd(num)
```

```python
#function with multiple parameters
def sum(a,b):
    """this function returns the sum of two numbers"""
    return a + b

sum(23, 345)
```

```python
#default parameter

def greet(name, msg = "Good Day!"):
    """this function greets a person with a message"""
    print(f"Hi {name}, {msg}")

greet("Shruti")
greet("Shruti", "Good evening! How are you?")
```

```python
# positional arguments

def sum(*args):
    """this function returns the sum of multiple numbers"""
    total = 0
    for num in args:
        total += num
    return total

sum(243, 463,1246, 2423, 12)
     
```

```python
#keyword arguments

def print_details(**kwargs):
    """this function prints the details of a person"""
    for key, value in kwargs.items():
        print(f"{key} : {value}")

print_details(name = "Shruti", age = 24, city = "Bangalore", country = "India")
```

```python
#positional and keyword arguments

def print_details(*args, **kwargs):
    """this function prints the details of a person"""
    for arg in args:
        print(f"positional argument: {arg}")
    for key, value in kwargs.items():
        print(f"{key} : {value}")

print_details(1, 34, "star", name = "Shruti", age = 24, city = "Bangalore", country = "India")

# you cannot have a positional argument after a keyword argument. It will give an error.
```
