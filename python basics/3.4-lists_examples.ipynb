```python
#real life examples

todo_list = ["study dsa", "complete gen ai section 2", "exercise", "read a book", "cook dinner"]

todo_list.append("call mom")  # Adding a new task to the list

todo_list.remove("exercise")  # Removing a completed task from the list

if "read a book" in todo_list:
    print("Don't forget to read your book today!")  # Reminder for a specific task  

print("Here is your to-do list for today:")
for task in todo_list:
    print(f"- {task}")  # Printing all tasks in the to-do list
```

```python
#create a list to store and calculate average grades of student

grades = [85, 90, 78, 92, 88]  # List of student grades

avg = sum(grades) / len(grades)
print(f"The average grade of the student is: {avg:.2f}")  # Displaying the average grade

max_grade = max(grades)
print(f"The maximum grade of the student is: {max_grade}")  # Displaying the maximum grade

min_grade = min(grades)
print(f"The minimum grade of the student is: {min_grade}")  # Displaying the minimum grade
```

```python
# using list to manage inventory 

inventory = ["apples", "bananas", "oranges", "grapes"]  # List of items in inventory

inventory.append("mangoes")  # Adding a new item to the inventory
inventory.remove("apples")  # Removing an item from the inventory

if "apples" in inventory:
    print("We have apples in stock!")  
if "oranges" in inventory:
    print("oranges available in stock!")  
else:
    print("Item not in stock.") 

print("inventory list -")

for fruit in inventory:
    print(f" - {fruit}")  # Printing all items in the inventory
```

```python
# collecting user feedback and appending it to a list

feedback = ["Great service", "friendly staff", "quick response time", "Excellent service"]  # List of user feedback

feedback.append("Could improve on delivery time")  # Adding new feedback to the list

positive_feedback = sum(1 for f in feedback if 'great' in f.lower() or 'excellent' in f.lower())
print(f"Number of positive feedbacks: {positive_feedback}")  # Counting positive feedbacks

print("Here is the feedback we received:")
for f in feedback:
    print(f" - {f}")  # Printing all feedback received
```
